### Itertools-
Itertools is a module to handle iterators.Iterators are the data type that can be used in for loopslike lists.
* product
* permutations
* combinations
* accumulate.
* groupby
* infinite iterators


### product

```python
from itertools import product
```

```python
#this will give a cartesian product-
a = [1,2]
b = [3,4]
prod = product(a,b)
print(list(prod))
```

### permutations

```python
from itertools import permutations
```

```python
#this will give all the possible permutations-
a = [1,2,3]
perm = permutations(a)
print(list(perm))
```

### combinations-

```python
from itertools import combinations
```

```python
# this will give all the possible combinations with desired length-
a =[1,2,3,4]
comb = combinations(a,2)
print(list(comb))
```

### accumulate-


```python
from itertools import accumulate
```

```python
# this gives the accumulated sums-
a = [1,2,3,4]
acc = accumulate(a)
print(list(acc))
```

```python
# by defualt it will plus. You can give other operations-
import operator
a = [1,2,3,4]
acc = accumulate(a,func = operator.mul)
print(list(acc))
```

### groupby

```python
from itertools import groupby
```

```python
def smaller_then_3(x):
    return x<3
a = [1,2,3,4,5]
grp_obj = groupby(a,key = smaller_then_3)
for key, value in grp_obj:
    print(key, list(value))
```

### infinite iterators

```python
from itertools import count,cycle,repeat
```

```python
for i in count(10):
    print(i)
    if i == 15:
        break
```

```python
# cycle will keep moving in alist again and again-
a =[1,2,3]
for i in cycle(a):
    print(i)
```

```python
# repeat- it repeats a particular value a defined number times.
a=[1,2,3]
for i in repeat(1,4):
    print(i)
```
